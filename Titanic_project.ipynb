```python
import seaborn as sns  
sns.countplot(df['survived'])      #shows survived and not survived
plt.title('Survived and Not Survived')
plt.xlabel('Survived')
plt.ylabel('Counts ')
plt.show()
```

```python
sns.countplot(df['sex'])      #shows no. of female and male
plt.title('Male and Female')
plt.xlabel('Sex')
plt.ylabel('Counts ')
plt.show()

```

```python
df=df.replace(to_replace =["female", "male"],  
                            value =[1,0])   #encoding male and female as 0,1
```

```python
df
```

```python
#deleting useless column
cols = ['boat', 'body', 'home.dest']
df = df.drop(cols, axis=1)
```

```python
df
```

```python

sns.countplot(df['age'])      #shows female and male
plt.title('age')
plt.xlabel('year')
plt.ylabel('Counts ')
plt.show()
```

```python
df['AgeRange'] = pd.cut(df['age'], [0, 15, 80], labels=['child', 'adult'])
df
```

```python
sns.catplot(x="AgeRange", y="survived", kind="bar", data=df)   #graph of age vs survived
```

```python
sns.countplot(df['AgeRange'])      #shows children and adult
plt.title('Child and Adult')
plt.xlabel('Age')
plt.ylabel('Counts ')
plt.show()
```

```python
df[['pclass', 'survived']].groupby(['pclass'], as_index=False).mean().sort_values(by='survived', ascending=False)
```

```python
df[["sex", "survived"]].groupby(['sex'], as_index=False).mean().sort_values(by='survived', ascending=False)
```

```python
df[["sibsp", "survived"]].groupby(['sibsp'], as_index=False).mean().sort_values(by='survived', ascending=False)
```

```python
df[["parch", "survived"]].groupby(['parch'], as_index=False).mean().sort_values(by='survived', ascending=False)
```
